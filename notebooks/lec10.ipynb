## Yield keyword
- Yield retains can retain the state of a function, so you can later iterate over a sequence of the same 'generator' process
- Function is a generator if it uses yield

```python
# This is an example of plotting some process, in this case process == sin(x), using yield to sequence over a range

# These are imports
# Under I am getting a function (sin) from the library math and then creating a reference
# to that function as (s). Not necessary, but I just wanted to showcase this feature
from math import sin as s
# This line below means that I'm importing the numpy then assigning np == numpy
# this allows me to use np.(LIB FUNCTION) instead of numpy.(LIB FUNCTION)
# We need np to use a range in our for loop and have it increment by a float number
# range() can only do steps by int so 1,2,3,... instead of 0.1,0.2,0.3,...
import numpy as np
import matplotlib.pyplot as plt

def process(x):
    return s(x)

def make_function(start, end):
    for i in np.arange(start, end, 0.01):
        yield process(i)

    plt.figure()

for i,j in enumerate(make_function(0, m.pi)):
    # print(f"i: {i}, j: {j}")
    plt.scatter(i, j)
```
